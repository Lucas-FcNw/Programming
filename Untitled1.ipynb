````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
dados = {
    "nome": [
        "Ana", "Bruno", "Carlos", "Daniela", "Eduardo",
        "Fernanda", "Gabriel", "Helena", "Igor", "Julia"
    ],
    "idade": [23, 35, 31, 29, 42, 37, 26, 34, 41, 28],
    "salario": [
        3200, 5400, 4800, 3900, 6200,
        5800, 3500, 5100, 6000, 4000
    ],
    "departamento": [
        "TI", "Financeiro", "TI", "Marketing", "Financeiro",
        "RH", "TI", "Marketing", "Financeiro", "RH"
    ]
}
````

````python
df = pd.DataFrame(dados)

df
````

````python
df.head()
````

````python
df.describe()
````

````python
# Funcionrios do departamento de TI
df_ti = df[df["departamento"] == "TI"]
df_ti
````

````python
# Pessoas com salrio acima de 5000
df_salario_alto = df[df["salario"] > 5000]
df_salario_alto
````

````python
# Salrio mdio por departamento
salario_medio = df.groupby("departamento")["salario"].mean()
salario_medio
````

````python
# Contagem de pessoas por departamento
contagem = df["departamento"].value_counts()
contagem
````

````python
sns.set_theme(style="whitegrid")
````

````python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="departamento", y="salario")
plt.title("Distribuio de salrios por departamento")
plt.show()
````

````python
plt.figure(figsize=(8, 5))
plt.bar(salario_medio.index, salario_medio.values)
plt.title("Salrio mdio por departamento")
plt.xlabel("Departamento")
plt.ylabel("Salrio mdio")
plt.show()

# Matplotib para quando precisa de um ajuste mais fino
````

````python
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x="idade",
    y="salario",
    hue="departamento"
)
plt.title("Idade vs Salrio")
plt.show()
````

## O que o Seaborn faz (e o Matplotlib NO faz sozinho)

Seaborn sabe **estatstica** e **contexto de dados**.

### Exemplo real

> Quero ver a distribuio de salrio por departamento

Voc no quer desenhar barras.  
Voc quer:

- mediana  
- quartis  
- outliers  
- comparao entre grupos  

Isso  **estatstica visual**.

 Seaborn entende isso.

```python
sns.boxplot(data=df, x="departamento", y="salario")
```

Com **1 linha**, ele decide:

- tipo de grfico correto  
- como agrupar  
- como calcular quartis  
- como desenhar outliers  

 O Matplotlib no sabe o que  departamento, salrio ou distribuio.
