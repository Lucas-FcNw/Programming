```python
from pyspark.sql import functions as F
from pyspark.sql.types import *

# Em um notebook real, o 'df' já viria carregado anteriormente.
# Aqui, assumimos que o df já existe no contexto de execução do validador.
```

```python
# Regra de Renda Alta
# O RMA diz que acima de 2000 é True. Vamos fazer igual.
df = df.withColumn("regra_renda_alta", F.when(F.col("renda") > 2000, True).otherwise(False))
```

```python
# Regra de Score Bom
# ATENÇÃO: O RMA oficial diz > 800.
# Aqui o desenvolvedor "errou" e colocou > 950. O validador TEM que pegar isso.
df = df.withColumn("regra_score_bom", F.when(F.col("score") > 950, True).otherwise(False))
```

```python
# Regra 3: Elegibilidade Premium (Uso de E)
# Estrutura: SE (condição 1 E condição 2) ENTÃO True SENÃO False
df = df.withColumn("regra_elegivel_premium", 
    F.when(
        (F.col("renda") > 3000) & (F.col("score") > 700), 
        True
    ).otherwise(False)
)
```

```python
# Regra 4: Bloqueio de Risco (Uso de OU)
# Estrutura: SE (condição 1 OU condição 2) ENTÃO True SENÃO False
df = df.withColumn("regra_bloqueio_risco", 
    F.when(
        (F.col("idade") < 18) | (F.col("score") < 400), 
        True
    ).otherwise(False)
)
```

```python
# Regra 5: Aprovação Automática (Mista)
# Estrutura: SE ((A e B) ou C) ENTÃO True SENÃO False
df = df.withColumn("regra_aprovacao_auto", 
    F.when(
        ( (F.col("uf") == "SP") & (F.col("score") > 800) ) | 
        (F.col("renda") > 15000),
        True
    ).otherwise(False)
)
```
