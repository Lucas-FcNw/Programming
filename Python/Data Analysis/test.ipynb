```python
from pyspark.sql import SparkSession
# 1. Criar ou obter a SparkSession
spark = SparkSession.builder \
    .appName("TestePySpark") \
    .master("local[*]") \
    .getOrCreate()

# 2. Criar dados de exemplo
data = [("Python", "Iniciante"), ("PySpark", "Intermediário"), ("Big Data", "Avançado")]
columns = ["Tecnologia", "Nivel"]

# 3. Criar o DataFrame
df = spark.createDataFrame(data, columns)

# 4. Exibir o conteúdo e o esquema
df.show()
df.printSchema()

# 5. Fechar a sessão
spark.stop()
```
